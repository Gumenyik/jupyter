```python
import pandas as pd
import seaborn as sbs
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

Завдання 1

```python
x = np.linspace(-10, 10, num = 1000)
y_norm = norm.pdf(x, loc = -1, scale = 3)
plt.plot(x, y_norm, color='k');
```

Завдання 2

```python
norm_rand_vals = norm.rvs(size = 1000, loc = -1, scale = 3)
```

Завдання 3

```python
plt.hist(norm_rand_vals, bins=20, density=True);
plt.plot(x, y_norm, color='k');
```

Завдання 4

```python
from scipy import stats
```

```python
kde_norm_pdf = stats.gaussian_kde(norm_rand_vals)
```

Завдання 5

```python
y_kde_norm = kde_norm_pdf(x)
```

Завдання 6

```python
plt.hist(norm_rand_vals, bins=20, density=True);
plt.plot(x, y_norm, color='k');
plt.plot(x, y_kde_norm);
```

Завдання 7

```python
norm_resamp_vals = kde_norm_pdf.resample(size = 1000)
norm_resamp_vals = norm_resamp_vals[0][:]
```

Завдання 8

```python
plt.hist(norm_rand_vals, bins=20, density=True);
plt.hist(norm_resamp_vals, bins=20, density=True, alpha= 0.5, color = 'red');
plt.plot(x, y_norm, color='k');
plt.plot(x, y_kde_norm);
```

Завдання 10

```python
from scipy.stats import rv_continuous
```

```python
my_pdf = lambda a: 0.3 * norm.pdf(a, -8, 0.5) + 0.3 * norm.pdf(a, -4, 2) + 0.4 * norm.pdf(a, 2, 1)
```

```python
class my_dist(rv_continuous):
    def _pdf(self, x):
        return my_pdf(x)
```

```python
md = my_dist()
x = np.linspace(-10, 5, num = 250)
y_norm = md.pdf(x)
plt.plot(x, y_norm, color='k');
```

```python
norm_rand_vals = md.rvs(size = 250)
```

```python
plt.hist(norm_rand_vals, bins=20, density=True);
plt.plot(x, y_norm, color='k');
```
